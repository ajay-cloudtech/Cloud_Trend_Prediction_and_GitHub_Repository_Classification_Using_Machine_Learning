```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet

# Load dataset
df = pd.read_csv("github_repos_filtered.csv")

# Convert 'created_at' to datetime format
df['created_at'] = pd.to_datetime(df['created_at'], format='mixed')

# List to store all forecast data
all_forecasts = []

# Function to train Prophet model, show monthly forecast summary, and save in one file
def forecast_trend(df, col_name, label):
    df_trend = df[df[col_name] == 1].groupby('created_at').size().reset_index(name='y')
    df_trend.rename(columns={'created_at': 'ds'}, inplace=True)

    if df_trend.empty:
        print(f"Skipping {label} - No data available.")
        return
    
    model = Prophet()
    model.fit(df_trend)

    future = model.make_future_dataframe(periods=365)
    forecast = model.predict(future)

    # Add category column and store forecast
    forecast["category"] = label
    all_forecasts.append(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'category']])

    # Aggregate predictions by month
    forecast['month'] = forecast['ds'].dt.to_period('M')
    monthly_forecast = forecast.groupby('month')['yhat'].sum().reset_index()

    # Print monthly forecast summary
    print(f"\n{label} Monthly Forecast Summary")
    for _, row in monthly_forecast.iterrows():
        print(f"{row['month']}: {round(row['yhat'])} projects")

    # Improve graph readability
    plt.figure(figsize=(12, 6))
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.plot(df_trend['ds'], df_trend['y'], 'bo-', label="Actual Data", markersize=3)
    plt.plot(forecast['ds'], forecast['yhat'], 'r-', label="Forecast", linewidth=2)
    plt.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='pink', alpha=0.3)
    plt.xlabel("Date", fontsize=12)
    plt.ylabel("Number of Projects", fontsize=12)
    plt.title(f"{label} Trend Forecast", fontsize=14)
    plt.legend()
    plt.show()

# Forecast overall cloud project trend
forecast_trend(df, "is_cloud_project", "Overall Cloud Projects")

# Forecast cloud platforms
for platform in ["AWS", "Azure", "GCP"]:
    forecast_trend(df, platform, f"{platform} Projects")

# Forecast top 3 cloud technologies
for tech in ["Docker", "Kubernetes", "Terraform"]:
    forecast_trend(df, tech, f"{tech} Projects")

# Forecast top 3 languages
for lang in ["JavaScript", "Python", "TypeScript"]:
    forecast_trend(df, lang, f"{lang} Projects")

# Save all forecasts in one CSV file
if all_forecasts:
    final_df = pd.concat(all_forecasts, ignore_index=True)
    final_df.to_csv("all_forecasts.csv", index=False)
    print("Saved all forecast results: all_forecasts.csv")
```
